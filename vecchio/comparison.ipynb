```python
import matplotlib.pyplot as plt
import pandas as pd

# Nomi dei file CSV (assicurati che corrispondano ai tuoi file)
cuda_base_csv_file = 'cuda_results_min_time.csv' # File CUDA Base
cuda_optimized_csv_file = 'cuda_optimized_results_min_time.csv' # File CUDA Ottimizzato
triton_csv_file = 'triton_results_min_time.csv' # File Triton Base

# Carica i dati CSV usando pandas
try:
    cuda_base_data = pd.read_csv(cuda_base_csv_file)
    cuda_optimized_data = pd.read_csv(cuda_optimized_csv_file)
    triton_data = pd.read_csv(triton_csv_file)
except FileNotFoundError as e:
    print(f"Errore: File CSV non trovato: {e}. Assicurati che tutti i file CSV siano nella stessa cartella di questo script.")
    exit()


# *** Grafico 1: Tempi di Esecuzione vs. Dimensione Matrice (con CUDA Ottimizzato) ***
plt.figure(figsize=(12, 7)) # Aumenta la dimensione per pi linee
plt.plot(cuda_base_data['Dimensione'], cuda_base_data['Tempo_CUDA_ms'], label='CUDA (Base)', marker='o', linestyle='--') # Stile linea modificato
plt.plot(cuda_optimized_data['Dimensione'], cuda_optimized_data['Tempo_CUDA_ms'], label='CUDA (Ottimizzato)', marker='s', linestyle='-') # Stile linea modificato
plt.plot(triton_data['Dimensione'], triton_data['Tempo_Triton_ms'], label='Triton (Base)', marker='x', linestyle=':') # Stile linea modificato
plt.xlabel('Dimensione Matrice (N)')
plt.ylabel('Tempo di Esecuzione (ms)')
plt.title('Confronto Tempi di Esecuzione: CUDA Base vs. CUDA Ottimizzato vs. Triton Base') # Titolo modificato
plt.legend()
plt.grid(True)
plt.xscale('log')
plt.yscale('log')
plt.savefig('tempi_cuda_base_optimized_vs_triton_base.png') # Nome file modificato
plt.show()


# *** Grafico 2a: Speedup di Triton rispetto a CUDA Base (gi esistente) ***
speedup_vs_cuda_base = cuda_base_data['Tempo_CUDA_ms'] / triton_data['Tempo_Triton_ms']
plt.figure(figsize=(12, 7)) # Aumenta la dimensione per pi chiarezza
plt.plot(cuda_base_data['Dimensione'], speedup_vs_cuda_base, label='Speedup vs. CUDA (Base)', marker='o', linestyle='--') # Nome legenda modificato
plt.axhline(y=1.0, color='r', linestyle='--', label='Speedup = 1 (Parit)')
plt.xlabel('Dimensione Matrice (N)')
plt.ylabel('Speedup (CUDA Base / Triton Base)') # Etichetta asse Y modificata
plt.title('Speedup di Triton rispetto a CUDA Base') # Titolo modificato
plt.legend()
plt.grid(True)
plt.xscale('log')
plt.yscale('log')
plt.savefig('speedup_triton_vs_cuda_base_updated.png') # Nome file modificato
plt.show()


# *** Grafico 2b: Speedup di Triton rispetto a CUDA Ottimizzato (NUOVO GRAFICO) ***
speedup_vs_cuda_optimized = cuda_optimized_data['Tempo_CUDA_ms'] / triton_data['Tempo_Triton_ms']
plt.figure(figsize=(12, 7)) # Aumenta la dimensione per pi chiarezza
plt.plot(cuda_optimized_data['Dimensione'], speedup_vs_cuda_optimized, label='Speedup vs. CUDA (Ottimizzato)', marker='s', linestyle='-') # Nome legenda modificato
plt.axhline(y=1.0, color='r', linestyle='--', label='Speedup = 1 (Parit)')
plt.xlabel('Dimensione Matrice (N)')
plt.ylabel('Speedup (CUDA Ottimizzato / Triton Base)') # Etichetta asse Y modificata
plt.title('Speedup di Triton rispetto a CUDA Ottimizzato') # Titolo modificato
plt.legend()
plt.grid(True)
plt.xscale('log')
plt.yscale('log')
plt.savefig('speedup_triton_vs_cuda_optimized.png') # Nome file modificato
plt.show()


print("Grafici aggiornati 'tempi_cuda_base_optimized_vs_triton_base.png', 'speedup_triton_vs_cuda_base_updated.png', e 'speedup_triton_vs_cuda_optimized.png' creati nella cartella corrente.")
```
